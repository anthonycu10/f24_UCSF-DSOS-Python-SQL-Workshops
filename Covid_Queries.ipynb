```python
# prompt: Download the code repo at: https://github.com/mdcollab/covidclinicaldata.git then build a dataframe with all the csv files in the data directory.

!git clone https://github.com/mdcollab/covidclinicaldata.git

import pandas as pd
import os
import glob

# Define the path to the data directory
data_dir = 'covidclinicaldata/data'

# Create an empty list to store the dataframes
dfs = []

# Iterate through all csv files in the data directory
for csv_file in glob.glob(os.path.join(data_dir, '*.csv')):
  # Read the csv file into a pandas dataframe
  df = pd.read_csv(csv_file)
  # Append the dataframe to the list
  dfs.append(df)

# Concatenate all dataframes into a single dataframe
combined_df = pd.concat(dfs, ignore_index=True)

# Print the combined dataframe
#print(combined_df)
```

```python
# prompt: run a sql query against the dataframe above, combined_df to show the count of positive and negative test results

from pandasql import sqldf

pysqldf = lambda q: sqldf(q, globals())

query = """
SELECT
  covid19_test_results,
  COUNT(*) AS result_count
FROM combined_df
WHERE covid19_test_results IS NOT NULL
GROUP BY covid19_test_results;
"""

result = pysqldf(query)
```

```python
# prompt: write a sql query that groups the count for  covid_19_test_results by high_risk_exposure_occupation and high_risk_interactions

query = """
SELECT
  covid19_test_results,
  high_risk_exposure_occupation,
  high_risk_interactions,
  COUNT(*) AS result_count
FROM combined_df
WHERE high_risk_exposure_occupation IS NOT NULL AND high_risk_interactions IS NOT NULL
AND covid19_test_results IS NOT NULL
GROUP BY covid19_test_results, high_risk_exposure_occupation, high_risk_interactions;
"""

result = pysqldf(query)
```

```python
# prompt: Using dataframe result: create a bar chart  using matplotlib showing high the positive and negative test count by high_rist_exposure_occupation

import matplotlib.pyplot as plt

# Group data by covid19_test_results and high_risk_exposure_occupation
grouped_data = result.groupby(['covid19_test_results', 'high_risk_exposure_occupation'])['result_count'].sum().unstack()

# Create the bar chart
grouped_data.plot(kind='bar', figsize=(10, 6))
plt.title('Positive and Negative Test Count by High-Risk Exposure Occupation')
plt.xlabel('COVID-19 Test Results')
plt.ylabel('Result Count')
plt.legend(title='High-Risk Exposure Occupation')
plt.show()
```
