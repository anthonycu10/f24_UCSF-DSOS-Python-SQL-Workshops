```python
query = """
SELECT
    covid19_test_results,
    high_risk_exposure_occupation,
    high_risk_interactions,
    cough, 
    cough_severity,
    fever, 
    sob, 
    sob_severity, 
    diarrhea, 
    fatigue, 
    headache,
    loss_of_smell, 
    loss_of_taste, 
    runny_nose, 
    muscle_sore,
    sore_throat
FROM
    df_covid
"""
```

```python
df_predictive = pysqldf(query)
```

```python
df_predictive
```

```python
df_predictive.describe(include="all")
```

```python
pysqldf("""
SELECT
    DISTINCT cough_severity
FROM df_predictive
""")
```

```python
# how many for each?
```

```python
pysqldf("""
SELECT cough_severity, COUNT(*) AS cs_count
FROM df_predictive
GROUP BY cough_severity
""")
```

```python
pysqldf("""
SELECT covid19_test_results, COUNT(*) AS cs_count
FROM df_predictive
GROUP BY covid19_test_results
""")
```

```python
pysqldf("""
SELECT 
    covid19_test_results, 
    high_risk_interactions,
    COUNT(*) AS cs_count
FROM df_predictive
WHERE high_risk_interactions == 1
GROUP BY 
    covid19_test_results,
    high_risk_interactions
""")
```
